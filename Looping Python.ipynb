```python
#1.

# Initialize variables
number = 1
sum_even = 0

# Use a while loop to go through numbers from 1 to 20
while number <= 20:
    if number % 2 == 0:  # Check if the number is even
        sum_even += number
    number += 1  # Increment to the next number

# Output the result
print("The sum of all even numbers between 1 and 20 is:", sum_even)
```

```python
#2.
# Prompt the user for input
number = int(input("Enter a number between 1 and 5: "))

# Keep asking until the number is within the range
while number < 1 or number > 5:
    print("Invalid input. Please try again.")
    number = int(input("Enter a number between 1 and 5: "))

# Output the valid input
print("Thank you! You entered:", number)
```

```python
#3.

import random
random_number = random.randint(1, 5)
guess = int(input("Guess a number between 1 and 5: "))

while guess != random_number:
    if guess < random_number:
        print("Too low! Try again.")
    else:
        print("Too high! Try again.")
    guess = int(input("Guess again: "))

    print("Congratulations! You guessed the correct number:", random_number)
```

```python
#4.

# Set the correct password
correct_password = "python123"

# Initialize the attempt counter
attempts = 0
max_attempts = 3

# Loop to allow the user up to 3 attempts
while attempts < max_attempts:
    # Ask for the user's password
    password = input("Enter the password: ")

    # Check if the password is correct
    if password == correct_password:
        print("Access granted!")
        break  # Exit the loop if the password is correct
    else:
        attempts += 1  # Increment the attempt counter
        print("Incorrect password.")

# Check if the user has reached the maximum number of attempts
if attempts == max_attempts:
    print("Access denied!")
```
